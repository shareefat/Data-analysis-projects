**DATA CLEANING**

```python
sales.describe()
```

```python
sales['month_year'] = sales['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
sales['month_year'].sample(10)
```

EXPLORATORY ANALISYS


```python
sales.head()
```

```python
# Group by 'month_year' and calculate the sum of the 'sales' column
sales_by_month = sales.groupby('month_year')['sales'].sum().reset_index()
```

```python
plt.figure(figsize=(15,6))
plt.plot(sales_by_month['month_year'],sales_by_month['sales'])
plt.xticks(rotation='vertical',size=10)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='month_year',y='sales',data=sales_by_month)
plt.xticks(rotation='vertical',size=10)
plt.show()
```

**MOST SALES**

```python
products_sales = pd.DataFrame(sales.groupby('product_name')['sales'].sum())
products_sales = products_sales.sort_values('sales',ascending=False)
```

```python
products_sales[:10]
```

```python
products_by_quantity = pd.DataFrame(sales.groupby('product_name')['quantity'].sum())
products_by_quantity_sorted = products_by_quantity.sort_values('quantity', ascending=False)
```

```python
products_by_quantity_sorted[:10]
```

**VISUALIZE MOST USED SHIP MODS**

```python
plt.figure(figsize=(10,8.5))
sns.countplot(sales['ship_mode'])
plt.show()
```

```python
cateory_by_profit = pd.DataFrame(sales.groupby(['category','sub_category'])['profit'].sum())
cateory_by_profit.sort_values(['profit'],ascending=False)
```
